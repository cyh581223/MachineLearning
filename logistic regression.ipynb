```python
count_0 = 0
count_1 = 0
for i in range(1000):
    if(np.random.randint(2) is 1):
        count_1+=1
    else:
        count_0+=1
print('count_0:',count_0)

para = 0.1 ** 0.5
y = np.zeros(1000)
x1 = para * np.random.randn(1000).reshape((-1,1))
x2 = para * np.random.randn(1000).reshape((-1,1))

for i in range(count_0,1000):
    x1[i]+=1
    x2[i]+=1
    y[i]+=1

#for i in range(0,1000):
#    tmp = np.random.randint(2)
#    x1[i] += tmp
#    x2[i] += tmp
#    y[i] += tmp
x = np.hstack([x1,x2])
#plt.plot(x1[:count_0],x2[:count_0],'ro')
#plt.plot(x1[count_0:],x2[count_0:],'bo')
#plt.show()
```

```python
model = LogisticRegression().fit(x,y)
#print(model.intercept_)
#print(model.coef_)
#plt.axis([-1.0, 2.5, -1.0, 2.5])
x_test = np.arange(-1.0, 2, 0.1).reshape(-1,1)
#plt.plot(x_test , (model.coef_[0][0] * x_test + (model.intercept_[0])) / (model.coef_[0][1] * -1 ) )
#plt.plot(x1,x2,'o')
#plt.show()
```
